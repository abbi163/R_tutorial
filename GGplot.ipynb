
### Installing the Package !!

```python
# Need to install just once !!
install.packages("tidyverse")
install.packages("gdata")
```

```python
# Tidyverse is for the plot visualization !!
library(tidyverse)
# gdata is for reading the csv files !!
library(gdata) 
```

    Reading CSV files !!

```python
mydata = read.csv("auto-mpg.csv")  # read csv file 
```

```python
mydata[ 0:5, 0:9 ]
```

    Here first define data = mydata and then define what kind of plot are we looking for, and define x and y axis !!

```python
ggplot(data = mydata) + geom_point(mapping = aes(x = mpg, y = displacement))
```

    If we select the color also accordingly, it will give a colorful plot. 

```python
ggplot(data = mydata) + geom_point(mapping = aes(x = mpg, y = displacement, color = cylinders))
```

    We can change the size also of the dots also

```python
ggplot(data = mydata) + geom_point(mapping = aes(x = mpg, y = displacement, size = cylinders))
```

    Mapping the class to alpha asthetics by changing the transparency !!

```python
ggplot(data = mydata) + geom_point(mapping = aes(x = mpg, y = displacement, alpha = cylinders))
```

    Mapping the class by changing the shape
    shape = class name. Remember class name should be a discrete value and it can deal with maximum 6 discrete values. so don't give more than that. 

```python
x = mydata[0:5 , 0:9]
ggplot(data = x) + geom_point(mapping = aes(x = mpg, y = displacement, shape = car.name))
```

```python
ggplot(data = mydata) + geom_point(mapping = aes(x = mpg, y = displacement)) + facet_wrap(~ cylinders, nrow = 2)
```
